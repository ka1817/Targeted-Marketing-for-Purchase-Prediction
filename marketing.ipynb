```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold
from joblib import dump

# Assuming `df` is your DataFrame
# Define feature columns and target column
X = df[['DemAffl', 'DemAge', 'LoyalSpend', 'LoyalTime', 'DemClusterGroup', 'DemGender', 'DemReg', 'DemTVReg', 'LoyalClass']]
y = df['TargetBuy']  # Replace with the actual target column name

# Define numeric and categorical columns
numeric_features = ['DemAffl', 'DemAge', 'LoyalSpend', 'LoyalTime']
categorical_features = ['DemClusterGroup', 'DemGender', 'DemReg', 'DemTVReg', 'LoyalClass']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define preprocessing for numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine them into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define classifiers
classifiers = {
    'RandomForest': RandomForestClassifier(random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'SVC': SVC(random_state=42)
}

# Hyperparameter tuning for each model
param_grids = {
    'RandomForest': {
        'clf__n_estimators': [50, 100],
        'clf__max_depth': [None, 10, 20],
        'clf__min_samples_split': [2, 5]
    },
    'GradientBoosting': {
        'clf__n_estimators': [50, 100],
        'clf__learning_rate': [0.05, 0.1, 0.2],
        'clf__max_depth': [3, 4, 5]
    },
    'SVC': {
        'clf__C': [0.1, 1, 10],
        'clf__gamma': ['scale', 'auto'],
        'clf__kernel': ['linear', 'rbf']
    }
}

# Create pipelines for each classifier
pipelines = {
    name: Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('clf', clf)
    ]) for name, clf in classifiers.items()
}

# Hyperparameter tuning and model training
best_models = {}

for name, pipeline in pipelines.items():
    print(f"Training {name}...")
    grid_search = GridSearchCV(pipeline, param_grids[name], cv=StratifiedKFold(5), n_jobs=-1, verbose=1)
    grid_search.fit(X_train, y_train)
    best_models[name] = grid_search.best_estimator_

    # Evaluate the best model
    y_pred = grid_search.best_estimator_.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"{name} - Accuracy: {accuracy:.4f}")
    
    # Save the best model
    dump(grid_search.best_estimator_, f"{name}_best_model.joblib")
    print(f"{name} model saved!")

print("Training and saving complete.")
```

```python
import mlflow
import mlflow.sklearn
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.model_selection import StratifiedKFold

# Assuming `df` is your DataFrame
# Define feature columns and target column
X = df[['DemAffl', 'DemAge', 'LoyalSpend', 'LoyalTime', 'DemClusterGroup', 'DemGender', 'DemReg', 'DemTVReg', 'LoyalClass']]
y = df['TargetBuy']  # Replace with the actual target column name

# Define numeric and categorical columns
numeric_features = ['DemAffl', 'DemAge', 'LoyalSpend', 'LoyalTime']
categorical_features = ['DemClusterGroup', 'DemGender', 'DemReg', 'DemTVReg', 'LoyalClass']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Define preprocessing for numeric and categorical data
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

# Combine them into a ColumnTransformer
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

# Define classifiers
classifiers = {
    'RandomForest': RandomForestClassifier(random_state=42),
    'GradientBoosting': GradientBoostingClassifier(random_state=42),
    'SVC': SVC(random_state=42)
}

# Hyperparameter tuning for each model
param_grids = {
    'RandomForest': {
        'clf__n_estimators': [50, 100],
        'clf__max_depth': [None, 10, 20],
        'clf__min_samples_split': [2, 5]
    },
    'GradientBoosting': {
        'clf__n_estimators': [50, 100],
        'clf__learning_rate': [0.05, 0.1, 0.2],
        'clf__max_depth': [3, 4, 5]
    },
    'SVC': {
        'clf__C': [0.1, 1, 10],
        'clf__gamma': ['scale', 'auto'],
        'clf__kernel': ['linear', 'rbf']
    }
}

# Create pipelines for each classifier
pipelines = {
    name: Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('clf', clf)
    ]) for name, clf in classifiers.items()
}

# Hyperparameter tuning and model training
best_models = {}

# Start an MLflow experiment
mlflow.set_experiment("Classifier_Models")

for name, pipeline in pipelines.items():
    print(f"Training {name}...")
    
    # Start a new MLflow run
    with mlflow.start_run() as run:
        grid_search = GridSearchCV(pipeline, param_grids[name], cv=StratifiedKFold(5), n_jobs=-1, verbose=1,scoring='f1')
        grid_search.fit(X_train, y_train)
        
        # Log parameters
        mlflow.log_params(grid_search.best_params_)
        
        # Log metrics
        y_pred = grid_search.best_estimator_.predict(X_test)
        accuracy = accuracy_score(y_test, y_pred)
        mlflow.log_metric("accuracy", accuracy)
        
        # Save the best model using MLflow
        mlflow.sklearn.log_model(grid_search.best_estimator_, f"{name}_best_model")
        
        best_models[name] = grid_search.best_estimator_

        print(f"{name} - Accuracy: {accuracy:.4f}")
        print(f"Logged {name} model in MLflow!")

print("Training, hyperparameter tuning, and model logging complete.")
```
